```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline

grcdf = pd.read_csv("../K1_mi10_GRC.csv")
grcdf.head()
```

```python
print(grcdf.shape)
print(grcdf.info())
```

```python
grcdf.describe()
```

```python
grcdf['Erm/Sci'].hist()  #replace the columns to see a histo for them
plt.xlabel('Values in Column')
plt.ylabel('Frequency')
plt.title('Histogram of Numeric Column')
plt.show()
```

```python
plt.scatter(grcdf['Erm/Sci'],grcdf['strain'],)
plt.xlabel('Erm/Sci')
plt.ylabel('strain')
plt.show()
```

```python
plt.scatter(grcdf['Scm/Po'].values,grcdf['strain'].values)
plt.xlabel('Scm/Po')
plt.ylabel('strain')
plt.show()
```

```python
# Define colors based on 'S' values
colors = plt.cm.viridis(grcdf['Scm/Po'] / max(grcdf['Scm/Po']))

# Create scatter plot
plt.scatter(grcdf['conf'], grcdf['strain'], c=colors, cmap='viridis')
cbar = plt.colorbar()
cbar.set_label('Scm/Po')

# Labeling and Title
plt.xlabel('confinement')
plt.ylabel('strain')
```
