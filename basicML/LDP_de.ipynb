```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline

ldpdf = pd.read_csv("../K1_mi10_LDP.csv")
ldpdf.head()
```

```python
print(ldpdf.shape)
print(ldpdf.info())
```

```python
ldpdf.describe()
```

```python
ldpdf['Erm/Sci'].hist()  #replace the columns to see a histo for them
plt.xlabel('Values in Column')
plt.ylabel('Frequency')
plt.title('Histogram of Numeric Column')
plt.show()
```

```python
plt.scatter(ldpdf['Erm/Sci'],ldpdf['strain'],)
plt.xlabel('Erm/Sci')
plt.ylabel('strain')
plt.show()
```

```python
plt.scatter(ldpdf['Scm/Po'].values,ldpdf['strain'].values)
plt.xlabel('Scm/Po')
plt.ylabel('strain')
plt.show()
```

```python
plt.scatter(ldpdf['X*'],ldpdf['strain'],)
plt.xlabel('X*')
plt.ylabel('strain')
plt.show()
```

```python
ldpdfs1 = ldpdf[ldpdf['Erm/Sci']==28.11]
plt.scatter(ldpdfs1['X*'],ldpdfs1['strain'],)
plt.xlabel('X*')
plt.ylabel('strain')
plt.show()
```

```python

# Define colors based on 'S' values
colors = plt.cm.viridis(ldpdf['Scm/Po'] / max(ldpdf['Scm/Po']))

# Create scatter plot
plt.scatter(ldpdf['X*'], ldpdf['strain'], c=colors, cmap='viridis')
cbar = plt.colorbar()
cbar.set_label('Scm/Po')
# Labeling and Title
plt.xlabel('X*')
plt.ylabel('strain')
```

```python

# Define colors based on 'S' values
colors = plt.cm.viridis(ldpdfs1['Scm/Po'] / max(ldpdfs1['Scm/Po']))

# Create scatter plot
plt.scatter(ldpdfs1['X*'], ldpdfs1['strain'], c=colors, cmap='viridis')
cbar = plt.colorbar()
cbar.set_label('Scm/Po')

# Labeling and Title
plt.xlabel('X*')
plt.ylabel('strain')
```
